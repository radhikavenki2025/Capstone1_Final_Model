```python
import importlib
import deploy
import pandas as pd  



# Load (simulating a fresh session)
loaded_model, feature_columns = deploy.load_model('student_mentalhealth.pkl')

# Predict on new student(s)
new_student = pd.DataFrame({
                            'age': [21],
                            'sleep_hrs': [3],
                            'media_hrs': [10],
                            'study_hrs': [2],
                            'part_time_job':['Yes'],
                            'att_percent':[10],
                            'diet_quality':['Poor'],
                            'exercise_fq':[1],
                            'internet_quality':['Poor'],
                            'extracurricular':['No'],
                            'exam_score':[20],
                            'gender':['Male'], 
                            'parent_ed':['High School']
                           })

new_student_prepro= deploy.preprocess_input( new_student,feature_columns)

prediction = loaded_model.predict(new_student_prepro)

print(prediction)

prediction = deploy.predict_student(loaded_model, new_student_prepro, feature_columns)

if prediction[0] == 0:
    print("⚠️ This student is AT RISK. Please follow up with a counsellor.")
else:
    print("✅ This student is NOT at risk.")
    
print(f"At-Risk Probability: {prediction}")
```

"The deployment pipeline functions correctly end to end. 
However, live testing confirmed the model's known limitation — it consistently predicts Not At Risk 
regardless of input values, consistent with the 0.22 recall finding from evaluation. 
This reinforces the conclusion that the dataset lacks sufficient signal in behavioral features to 
distinguish at-risk students reliably. A stronger dataset (e.g. GLOBEM passive sensing data) would be 
required for a production-ready tool."
