```python
import importlib
import deploy
import pandas as pd  



# Load (simulating a fresh session)
loaded_model, feature_columns = deploy.load_model('student_mentalhealth1.pkl')

# Predict on new student(s)
new_student = pd.DataFrame({
                            'age': [20],
                            'sleep_hrs': [4],
                            'media_hrs': [6],
                            'study_hrs': [3],
                            'part_time_job':['No'],
                            'att_percent':[60],
                            'diet_quality':['Good'],
                            'exercise_fq':[4],
                            'internet_quality':['Good'],
                            'extracurricular':['Yes'],
                            'exam_score':[60],
                            'gender':['Male'], 
                            'parent_ed':['Bachelor']
                           })

new_student_prepro= deploy.preprocess_input( new_student,feature_columns)

prediction = loaded_model.predict(new_student_prepro)

print(prediction)

prediction = deploy.predict_student(loaded_model, new_student_prepro, feature_columns)

pred, at_risk_prob, not_at_risk_prob = deploy.predict_new(loaded_model, new_student_prepro)

print(f"Prediction: {'Student Mental Health is At-Risk' if pred == 0 else 'Student Mental Health is Not At-Risk'}")
print(f"At-Risk Probability:     {at_risk_prob:.0%}")
print(f"Not At-Risk Probability: {not_at_risk_prob:.0%}")
```
